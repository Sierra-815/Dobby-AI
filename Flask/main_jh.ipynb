```
%cd /content/drive/Shareddrives/KPMG_Ideation/Dobby-AI/Flask
app = Flask(__name__)
run_with_ngrok(app)

basedir = '/content/drive/Shareddrives/KPMG_Ideation/Dobby-AI/Flask' 
upload_dir = os.path.join(basedir, 'uploads')  # basedir 의 uploads 에 파일 저장

%cd /content/drive/Shareddrives/KPMG_Ideation/Dobby-AI/Python_Modules
import make_tables

with open(os.path.join(upload_dir, 'faq_query.json')) as f:
    query_faq=json.load(f)
#################################################################################################################

admin = Admin(name='Uploaded Files')
admin.init_app(app)
dropzone = Dropzone(app)
admin.add_view(FileAdmin(upload_dir, name='FILES')) 
app.config['DROPZONE_ALLOWED_FILE_CUSTOM'] = True
app.config['DROPZONE_ALLOWED_FILE_TYPE'] = 'audio/*, .mp3, .wav, .m4a'
# app.config['DROPZONE_REDIRECT_VIEW']='result'


#############################################################################
@app.route("/", methods=['GET', 'POST'])
def upload():
    if request.method == 'POST':
        f = request.files.get('file')
        f.save(os.path.join(upload_dir, f.filename))
    return render_template('homepage.html')


@app.route("/result", methods=['GET', 'POST'])
def result():
    %cd /content/drive/Shareddrives/KPMG_Ideation/Dobby-AI/Flask/uploads
    f = open("script_agenda.txt", "r")
    full = f.readline()
    f.close()
    mtl = make_tables.MAKE_TABLES(query_faq['query_dict_agenda'], full,  types="Agenda")
    mtl.get_table()
    with open(os.path.join(basedir, "templates/result.html"), "r", encoding="UTF-8") as file:
            result_1 = file.read()
    with open(os.path.join(basedir, "templates/result_0.html"), "r", encoding="UTF-8") as file:
            result_2 = file.read()
    with open(os.path.join(upload_dir, "Output.txt"), 'r', encoding="UTF-8") as file:
        txt = file.read()
    result_1 += txt
    result_1 += result_2
    with open(os.path.join(basedir, "templates/final_result.html"), "w", encoding="UTF-8") as file:
            file.write(result_1)
    return render_template('final_result.html')    


@app.route("/result_interview", methods=['GET', 'POST'])
def result1():
    if request.method == 'POST':
        f = request.files.get('file')
        f.save(os.path.join(upload_dir, f.filename))
    return render_template('result1.html')
    

@app.route("/result_idea", methods=['GET', 'POST'])
def result2():
    %cd /content/drive/Shareddrives/KPMG_Ideation/Dobby-AI/Flask/uploads
    f = open("script_idea.txt", "r")
    full = f.readline()
    f.close()
    mtl = make_tables.MAKE_TABLES(query_faq['query_dict_idea'], full,  types="Idea")
    mtl.get_table()
    with open(os.path.join(basedir, "templates/result2.html"), "r", encoding="UTF-8") as file:
            result_1 = file.read()
    with open(os.path.join(basedir, "templates/result_1.html"), "r", encoding="UTF-8") as file:
            result_2 = file.read()
    with open(os.path.join(upload_dir, "Output.txt"), 'r', encoding="UTF-8") as file:
        txt = file.read()
    result_1 += txt
    result_1 += result_2
    with open(os.path.join(basedir, "templates/final_result.html"), "w", encoding="UTF-8") as file:
            file.write(result_1)
    return render_template('final_result.html')      
##############################################################################
if __name__ == '__main__':
    app.run()
```
